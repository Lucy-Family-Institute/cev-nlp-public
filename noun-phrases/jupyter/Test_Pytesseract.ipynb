```python
# conda install -c plotly chart-studio

import pytesseract
from pdf2image import convert_from_path
import glob
from root import DIR_INPUT

# pdfs = glob.glob(r"yourPath\*.pdf")
pdfs = ['050453121002-201500909-01 Turbo 09 julio 2018.pdf', '5326_sentencia Drummond Los Naranjos.pdf']
for pdf_path in pdfs:
    path = DIR_INPUT + '/tierras/'
    pages = convert_from_path(path+pdf_path, 500)

    for pageNum,imgBlob in enumerate(pages):
        text = pytesseract.image_to_string(imgBlob,lang='eng+spa+fra')

        with open(f'{pdf_path[:-4]}_page{pageNum}.txt', 'w') as the_file:
            the_file.write(text)
            print(text)

        if pageNum %  2 == 0:
            break
```
